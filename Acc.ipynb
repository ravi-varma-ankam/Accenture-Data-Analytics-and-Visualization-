```python
data = pd.read_csv('file.csv')
data
```

```python
# Finding null values
data.isnull().sum().sum()
```

```python
data.head(5)
```

```python
# columns
data.columns
```

```python
#dropping unnecessary columns
data.drop(['Unnamed: 0.1', 'Unnamed: 0_x', 'Unnamed: 0_y', 'Unnamed: 0'], axis=1, inplace=True)
data
```

```python
data
```

```python
#Changing capitals to lowercase
data['Category'] = data['Category'].str.lower()
data['Category']= data['Category'].replace('"', '', regex=True)
```

```python
# Number of different categories
data.groupby('Category').Category.count().sort_values(ascending = False).count()
```

```python
# Number of different categories
data.groupby('Category').Category.count().sort_values(ascending = False)
```

```python
# top 5 in categories according to score
top5 = data.groupby('Category').sum('Score').sort_values('Score', ascending = False).head(5)
top5
```

```python
data.head(3)
```

```python
#convert Datetime format from object to Date
data['Datetime'] = pd.to_datetime(data['Datetime'])
```

```python
data.groupby(data.Datetime.dt.month).Datetime.count()
#there are more posts in january month
```
